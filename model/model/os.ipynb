```python
import os
import requests
from serpapi import GoogleSearch
from PIL import Image
from io import BytesIO

# üîπ Replace with your actual SerpAPI key
SERPAPI_KEY = "d800049270738d4289215c622c19a5373cf1092278332af21f0c43799ac6e46d"

# üîπ Dictionary mapping dish names to their corresponding save directories
dishes = {
    "ŸÉÿ¥ÿ±Ÿä": r"D:/model/scripting data/koshary",
    "ŸÖÿ≠ÿ¥Ÿä Ÿàÿ±ŸÇ ÿπŸÜÿ®": r"D:/model/scripting data/wara2 3enab",
    "ŸÉŸÅÿ™ÿ© ŸÖÿ¥ŸàŸäÿ©": r"D:/model/scripting data/kofta",
    "ŸÅÿ™ÿ© ÿ®ÿßŸÑŸÑÿ≠ŸÖÿ©": r"D:/model/scripting data/fatet la7ma",
    "ŸÖŸÖÿ®ÿßÿ± ŸÖÿ≠ÿ¥Ÿä": r"D:/model/scripting data/mombaar ma7shy",
    "ÿ∑ÿßÿ¨ŸÜ ÿ®ÿßŸÖŸäÿ© ÿ®ÿßŸÑŸÑÿ≠ŸÖÿ©": r"D:/model/scripting data/bamya bella7ma",
    "ŸÖŸÑŸàÿÆŸäÿ©": r"D:/model/cripting data/molo5eyet araneb",
    "ÿµŸäŸÜŸäÿ© ÿ®ÿ∑ÿßÿ∑ÿ≥ ÿ®ÿßŸÑŸÑÿ≠ŸÖÿ©": r"D:/model/scripting data/batates bella7ma"
}

# üîπ Create directories if they don't exist
for path in dishes.values():
    os.makedirs(path, exist_ok=True)

def download_image(url, save_dir, count):
    """Downloads and saves an image from a URL."""
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))
        img_format = img.format if img.format else "jpg"
        img.save(os.path.join(save_dir, f"image_{count}.{img_format.lower()}"))
        print(f"‚úÖ Saved {count} in {save_dir}")
    except Exception as e:
        print(f"‚ùå Failed to download {url}: {e}")

def fetch_images(search_term, save_dir):
    """Fetches images for a given search term and downloads them."""
    params = {
        "q": search_term,
        "tbm": "isch",
        "ijn": 0,
        "num": 100,  # Fetch 100 images per request
        "api_key": SERPAPI_KEY
    }

    count = 0
    page = 0

    while count < 500:  # Fetch up to 1000 images
        params["ijn"] = page
        search = GoogleSearch(params)
        results = search.get_dict()
        images = results.get("images_results", [])

        if not images:
            print(f"‚ö†Ô∏è No more images found for {search_term}.")
            break

        for img in images:
            if count >= 500:
                break
            download_image(img["original"], save_dir, count)
            count += 1

        page += 1  # Move to the next page

# üîπ Run the function for each dish
for dish, path in dishes.items():
    print(f"üì∏ Downloading images for: {dish}")
    fetch_images(dish, path)

print("üéâ All images downloaded successfully!")
```
